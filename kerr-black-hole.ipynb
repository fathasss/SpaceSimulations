```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d as Axes3D

#Kara delik paramterleri
M = 1.0 #Kütle
a = 0.7 #spin parametresi (|a| < M olmalı)

# Küresel koordinatlar (θ ve φ)
theta = np.linspace(0, np.pi, 100)
phi = np.linspace(0, 2 * np.pi, 100)
theta, phi = np.meshgrid(theta, phi)

# Ergosfer yarıçapı (Boyut: θ'ya bağlı)
r_ergosphere = M + np.sqrt(M**2 - (a**2) * np.cos(theta)**2)

# Kartezlen koordinatlara çevirme
x = r_ergosphere * np.sin(theta) * np.cos(phi)
y = r_ergosphere * np.sin(theta) * np.sin(phi)
z = r_ergosphere * np.cos(theta)

# 3D çizim
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis', edgecolor='k', alpha=0.9)

ax.set_title("Kerr BlackHole Ergosphere (a = {:.2f})".format(a), fontsize=14)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_box_aspect([1, 1, 1])  # Eş oranlı eksenler

plt.show()
```
